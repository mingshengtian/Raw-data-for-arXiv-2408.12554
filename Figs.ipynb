# Raw data and figures

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.rcParams['font.size'] = 20
plt.rcParams['font.family'] = 'Times New Roman'
```

## Fig.1

```python
# data loading
listmode=np.array([211,22,31,4])
ave=np.zeros((4,6))
err=np.zeros((4,6))
for index,modeindex in enumerate(listmode):
    # read cvs file
    df = pd.read_csv('data/stellar-'+str(modeindex)+'modes-15000.csv', header=None)
    list= df.values

    # remove the data which is smaller than 1e-2 by ppt criteria
    listp0=list[4,:]
    listp1=np.where(listp0 > 1e-2)[0]
    list=list[:,listp1]

    # calculate the ave and error with 1000 data each time

    for ii in range (6):
        results = np.zeros(10)
        # Divide the first 10000 data into 10 parts, each part has 1000 data
        for i in range(10):
            # Get the data for the current part
            list0 = list[ii,i*1000:(i+1)*1000]

            alist1 = np.where(list0 > 1e-2)[0]
            alist2 = np.where(list0 <= 1e-2)[0]

            # Add the result to the results list
            results[i] = len(alist1) / len(list0)


        ave[index,ii] = np.mean(results)
        err[index,ii] = np.std(results)
```

```python
# plot figure based on the ave and err
x=np.linspace(1,4,4)


y1=ave[0:4,0]
y2=ave[0:4,1]
y3=ave[0:4,2]
y4=ave[0:4,3]

error1=err[0:4,0]
error2=err[0:4,1]
error3=err[0:4,2]
error4=err[0:4,3]


# Define the width of the bars
bar_width = 0.16
bar_width1=bar_width+0.02

# Create the positions of the bars on the x-axis
r1 = np.arange(len(y3))
r2 = [x+ bar_width1 for x in r1]
r3 = [x + bar_width1 for x in r2]
r4 = [x+ bar_width1 for x in r3]


plt.figure(figsize=(6, 4))
# Create the bar plot
plt.bar(r1, y1, width=bar_width, color='#374f95',label='1st')
plt.bar(r2, y2, width=bar_width, color='#699eba',label='2nd')
plt.bar(r3, y3, width=bar_width, color='#efb069',label='3rd')
plt.bar(r4, y4, width=bar_width, color='#dd5c3c',label='4th')



# Create the error bar plot
plt.errorbar(r1, y1, yerr=error1, markersize=2, fmt='none', ecolor='black', elinewidth=2, capsize=2)
plt.errorbar(r2, y2, yerr=error2, markersize=2, fmt='none',  ecolor='black', elinewidth=2, capsize=2)
plt.errorbar(r3, y3, yerr=error3, markersize=2, fmt='none',  ecolor='black', elinewidth=2, capsize=2)
plt.errorbar(r4, y4, yerr=error4, markersize=2, fmt='none',  ecolor='black', elinewidth=2, capsize=2)




plt.yticks([0.9,0.92,0.94,0.96,0.98,1])

plt.xticks([r + 1.66*bar_width for r in range(len(y3))], ['(2,1,1)', '(2,2)', '(3,1)','(4)'])

plt.tick_params(axis='both', direction='in')

# Set border width
for spine in plt.gca().spines.values():
    spine.set_linewidth(2)

plt.ylim(0.89, 1.02)

# Remove tick labels.
plt.gca().set_xticklabels([])
plt.gca().set_yticklabels([])

# Set border width
for spine in plt.gca().spines.values():
    spine.set_linewidth(2)

plt.show()
```
